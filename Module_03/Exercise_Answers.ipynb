```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/dataset.csv')
```

```python
print('Mean: ' + str(df['WHOIS_STATEPRO'].value_counts().mean()))
print('Mode: ' + str(df['WHOIS_STATEPRO'].value_counts().mode()))
print('Median: ' + str(df['WHOIS_STATEPRO'].value_counts().median()))
print('Range: ' + str(df['WHOIS_STATEPRO'].value_counts().max() - df['WHOIS_STATEPRO'].value_counts().min()))
print('Standard Deviation: ' + str(df['WHOIS_STATEPRO'].value_counts().std()))
print('Variance: ' + str(df['WHOIS_STATEPRO'].value_counts().var()))
```

```python
df['WHOIS_STATEPRO'].value_counts().hist()
```

```python
df['WHOIS_STATEPRO'].value_counts().nlargest(10).plot.bar()
```

```python
df['WHOIS_STATEPRO'].value_counts().nsmallest(10).plot.bar()
```
